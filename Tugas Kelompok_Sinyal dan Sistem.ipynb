### Simulasi Skema Analog Ke Digital Converter Dari Proses Sampling

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_conversion(input_signal, num_of_bits):
    output_signal = np.zeros(len(input_signal))
    threshold = 0

    for i in range(num_of_bits):
        if i % 2 == 0:
            threshold += 0.5 / (2 ** (i + 1))
        else:
            threshold -= 0.5 / (2 ** (i + 1))

    for i in range(len(input_signal)):
        if input_signal[i] >= threshold:
            output_signal[i] = 1
        else:
            output_signal[i] = 0

    return output_signal

# Generate a sequence of sine waves with random frequencies
input_signal = np.zeros(100)
for i in range(10):
    freq = np.random.uniform(0.5, 2)
    phase = np.random.uniform(0, 2*np.pi)
    input_signal += np.sin(2*np.pi*freq*np.arange(100)/100 + phase)

# Simulate conversion
num_of_bits = 8
output_signal = simulate_conversion(input_signal, num_of_bits)

# Visualize the signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(input_signal)
plt.title('Input Signal')
plt.subplot(2, 1, 2)
plt.plot(output_signal)
plt.title('Output Signal')
plt.show()
```

### Simulasi Skema Analog Ke Digital Converter Dari Proses Decoding

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

# Create function for digital conversion
def digital_conversion(decimal_number, num_of_bits):
    binary_number = bin(decimal_number)[2:] #convert to binary
    if len(binary_number) < num_of_bits: # add leading zeros
        binary_number = '0'*(num_of_bits - len(binary_number)) + binary_number
    return binary_number

# Simulate the process
def simulate_conversion(input_signal, num_of_bits):
    # Step 1: Apply a threshold (decoder) to convert the input signal into a digital value.
    threshold = np.mean(input_signal) # compute the threshold
    digital_values = [1 if sample > threshold else 0 for sample in input_signal] # apply threshold

    # Step 2: Perform digital conversion by decoding the digital values.
    binary_numbers = [digital_conversion(digital_value, num_of_bits) for digital_value in digital_values]
    decimal_numbers = [int(binary_number, 2) for binary_number in binary_numbers]

    return decimal_numbers

# Test the simulation
# Define input signal
input_signal = np.random.normal(0, 1, 100)

# Number of bits
num_of_bits = 4

# Simulate conversion
output_signal = simulate_conversion(input_signal, num_of_bits)

# Visualize the signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(input_signal)
plt.title('Input Signal')
plt.subplot(2, 1, 2)
plt.plot(output_signal)
plt.title('Output Signal')
plt.show()
```
